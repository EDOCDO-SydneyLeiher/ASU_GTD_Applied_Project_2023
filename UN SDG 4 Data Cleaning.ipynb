Data Cleaning

```python
# Import necessary libraries
import pandas as pd

# Load the Excel file into a DataFrame
file = '4.2.2.xlsx'
df = pd.read_excel(file, sheet_name='4.2.2')

# Display the first few rows of the DataFrame
df.head()

## Data Filtering

### Define Values to Keep in 'Column1'

values_to_keep = ['Algeria', 'Bahrain', 'Egypt', 'Iran (Islamic Republic of)', 'Iraq', 'Jordan', 'Kuwait', 'Lebanon',
                  'Libya', 'Morocco', 'Oman', 'Palestine', 'Qatar', 'Saudi Arabia', 'Syrian Arab Republic', 'Tunisia',
                  'United Arab Emirates', 'Yemen']

# Use boolean indexing to filter rows
df_filtered = df[df['Location'].isin(values_to_keep)]
df_filtered.head()
```

Drop Columns with all NaN Values

```python
# Drop columns with all NaN values
df_cleaned = df_filtered.dropna(axis=1, how='all')
df_cleaned.head()
```

Specify and Select Columns to Keep


```python
# Specify the columns to keep
columns_to_keep = ['Location', 'Period', 'Dim1', 'Value']

# Select and keep only the specified columns
df_selected = df_cleaned[columns_to_keep]
df_selected.head(10)
```

Rename Columns

```python
# Rename columns for clarity
df_selected = df_selected.rename(columns={'Location': 'Country', 'Period': 'Year'})
df_selected.head()
```

Analyzing Years and Countries


```python
# Function to find earliest and most recent years for each country
def earliest_and_most_recent_years(years_and_countries):
    country_years = {}
    
    for year, country in years_and_countries:
        if country not in country_years:
            country_years[country] = {"earliest": None, "most_recent": None}
        
        if country_years[country]["earliest"] is None or year < country_years[country]["earliest"]:
            country_years[country]["earliest"] = year
        
        if country_years[country]["most_recent"] is None or year > country_years[country]["most_recent"]:
            country_years[country]["most_recent"] = year
    
    return country_years

# Create a list of tuples (Year, Country)
year_country_data = list(zip(df_selected['Year'], df_selected['Country']))

# Find earliest and most recent years for each country
result = earliest_and_most_recent_years(year_country_data)

# Display the results
for country, years in result.items():
    print(f"Country: {country}, Earliest Year: {years['earliest']}, Most Recent Year: {years['most_recent']}")
```

Filterting Data

```python
# Add columns for earliest and most recent years to the DataFrame
df_selected['earliest_year'] = [result[country]['earliest'] for country in df_selected['Country']]
df_selected['most_recent_year'] = [result[country]['most_recent'] for country in df_selected['Country']]

# Function to filter rows
def filter_rows(row):
    return row['Year'] == row['earliest_year'] or row['Year'] == row['most_recent_year']

# Apply the filter function
filtered_df = df_selected[df_selected.apply(filter_rows, axis=1)]
filtered_df = filtered_df.reset_index(drop=True)
filtered_df.head(100)
```

Adding Year Type

```python
# Function to determine the type of year
def determine_year_type(row):
    if row['Year'] == row['earliest_year']:
        return 'Base'
    elif row['Year'] == row['most_recent_year']:
        return 'Most Recent'
    else:
        return None

# Add a 'Year_Type' column to the DataFrame
filtered_df['Year_Type'] = filtered_df.apply(determine_year_type, axis=1)
filtered_df.head()
```

Formatting Values

```python
# Format 'Value' column as a percentage
filtered_df['Value'] = (filtered_df['Value']).astype(str) + '%'
filtered_df.head()
```

Exporting to csv

```python
# Define the path for the output CSV file
output_csv_path = 'output_4_2_2.csv'

# Use the to_csv() method to export the DataFrame to a CSV file
filtered_df.to_csv(output_csv_path, index=False)

print(f"DataFrame has been exported to {output_csv_path}")
```
